```python
import xlrd
import csv

def csv_from_excel(file_name, date_string):
    wb = xlrd.open_workbook(f'../sample_data/{file_name.lower()}_{date_string}.xlsx')
    sh = wb.sheet_by_name(f'{file_name.upper()}_{date_string}')
    your_csv_file = open(f'../chosen_data/cross_walk/{file_name.lower()}_{date_string}.csv', 'w')
    wr = csv.writer(your_csv_file, quoting=csv.QUOTE_ALL)

    for rownum in range(sh.nrows):
        wr.writerow(sh.row_values(rownum))

    your_csv_file.close()
```
